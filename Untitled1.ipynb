```python
class Points(object):
    def __init__(self, x, y): 
        self.x = x 
        self.y = y 
    def print_point(self): 
        print('x=', self.x, ' y=', self.y) 
p1 = Points("A", "B") 
p1.print_point()
```

```python
class WeatherData:
    def __init__(self, temp_c, humidity):
        self.temp_c = temp_c
        self.humidity = humidity
    def display(self):
        print(f"Temperature: {self.temp_c}°C, Humidity: {self.humidity}%")
today = WeatherData(22, 45)
today.temp_c = 25
today.display()
```
